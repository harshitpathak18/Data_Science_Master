### Q1. What is Flask Framework? What are the advantages of Flask Framework?

### Flask Framework-
    Flask is a micro web framework written in Python that allows developers to quickly build web applications.It is a minimalistic framework, meaning it provides only the basic tools needed to build web applications, without any additional features or functionality that may not be needed.This makes Flask a great choice for building simple      web applications or for rapid prototyping. Flask is also very flexible, allowing developers to add additional functionality as needed through the use of extensions.

### Advantages of Flask Framework-

1. **Scalable**:Due to its status as a microframework, Flask may use it to quickly grow technological projects like web apps. If you want to make an app that starts out modest but has the potential to grow quickly and in areas you haven’t completely explored, this is a great option.



2. **Flexible**:This is the most crucial aspect of Flask and one of its key benefits. To paraphrase one of the Zen of Python’s maxims, simplicity can move about and rearrange more easily than complexity.



3. **Negotiation is simple**:To enable web developers to concentrate entirely on coding without becoming bogged down, navigational simplicity, like Django, is essential. The microframework’s core concepts are straightforward, saving web developers time and effort while giving them more control over the functionality of their code.


4. **Lightweight**:When we discuss a tool’s or framework’s design, we mean how few parts need to build and put back together, and how it doesn’t rely on a lot of extensions to function. Web developers have some control over using this method.

### Q3. What is App routing in Flask? Why do we use app routes?

    App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler.To bind a function to an URL path we use the app.route decorator.

### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”route to show the following details:
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999

```python
from flask import Flask

app=Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<center><h1>Welcome to ABC Corporation</h1></center>"

@app.route("/")
def details():
    html="""<h3>Company Name : ABC Corporation</h3>
            <h3>Location: India</h3>
            <h3>Contact Detail: 999-999-9999</h3>"""
    return html

if __name__=="__main__":
    app.run()
```

### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. The url_for() function generates a URL for a given endpoint and any arguments provided. This function is useful because it ensures that URLs are always generated according to the rules of the application, even if the URL structure is changed.


```python
# Example code

from flask import Flask, redirect, url_for


app=Flask(__name__)

@app.route("/admin")
def admin():
    return "<h1 align='center'>Admin Page</h1>"


@app.route("/student")
def student():
    return "<h1 align='center'>Student Page</h1>"


@app.route("/user/<name>")
def User(name):
    if name.lower()=="student":
        return redirect(url_for('student'))   #url_for(function_name)
    
    if name.lower()=='admin':
        return redirect(url_for("admin"))
    
if __name__=="__main__":
    app.run()
```

#### In above code, if we enter url http://127.0.0.1:5000/user/student then this will be redirected to http://127.0.0.1:5000/student and if entered url is http://127.0.0.1:5000/user/admin then this will be redirected to http://127.0.0.1:5000/admin
